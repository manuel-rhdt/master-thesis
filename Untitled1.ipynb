```python
import numpy
import accelerate
import gillespie.configuration

conf = gillespie.configuration.get("configuration.toml")
sig_network, res_network = gillespie.configuration.read_reactions(conf)

count = 100
length = 100
timestamps = numpy.zeros((count, length), dtype=numpy.double)
trajectory = numpy.zeros((count, 1, length), dtype=numpy.uint32)
reaction_events = numpy.zeros((count, length - 1), dtype=numpy.uint32)

accelerate.simulate_trajectories(timestamps, trajectory, reaction_events, sig_network)
```

```python
averaging_signals = []
for i in range(count_avrg_signals // batch):
    sig = delayed(simulate)(
        batch, length, sig_network, initial_value=50, pure=False
    )
    sig = ta.from_delayed(sig, batch, length)
    averaging_signals.append(sig)
averaging_signals = ta.concatenate(averaging_signals)
averaging_signals
```

```python
signals = []
responses = []
for i in range(20_000 // batch):
    sig = delayed(simulate)(
        batch, length, sig_network, initial_value=50, pure=False
    )
    res = delayed(simulate)(
        batch, length, res_network, ext_trajectory=sig, initial_value=50, pure=False
    )
    sig = ta.from_delayed(sig, batch, length)
    res = ta.from_delayed(res, batch, length)
    signals.append(sig)
    responses.append(res)

signals = ta.concatenate(signals)
responses = ta.concatenate(responses)
```

```python
me = marginal_likelihood(traj_lengths, averaging_signals, responses, res_network)
me
```

```python
me = -da.reduction(
        me, chunk=logsumexp, aggregate=logsumexp, axis=1, dtype=numpy.double
    ) + numpy.log(me.shape[1])
me
```

```python
data = me.compute()
```
