```python
import numpy
import json
import matplotlib
from matplotlib import pyplot

import seaborn as sns
sns.set()
pyplot.rcParams['figure.dpi'] = 200
```

```python
data = json.load(open('histogram.json'));
```

```python
ds = numpy.array(data['single'])
dm = numpy.array(data['marginal'])
```

```python
# maxv = numpy.max(dm, axis=0, keepdims=True)
maxv = ds[numpy.newaxis, :]
exp_ds = numpy.exp(ds - maxv[0])
exp_dm = numpy.exp(dm - maxv)
```

```python
for i in [50, 100, 150, 200]:
    ax = sns.distplot(exp_dm[:,i], label=f"$t = {i}$")
ax.set(xlabel='scaled likelihood', ylabel='probability density')
ax.axvline(x=1, color='red', label="$\mathrm{P}(X_i|S_i)$")
ax.legend()
ax.set(xlim=(-0, 2.5))
pyplot.savefig('histogram.pdf')
```

```python
for i in [50, 100, 200, 300, 500]:
    ax = sns.distplot(dm[:,i] - maxv[:,i], label=f"$t = {i}$")
ax.axvline(x=0, color='red', label="$\ln\;\mathrm{P}(X_i|S_i)$")
ax.set(xlabel='log likelihood', ylabel='probability density')
ax.legend()
ax.set(xlim=(-15, 2))
pyplot.savefig('histogram.pdf')
```
