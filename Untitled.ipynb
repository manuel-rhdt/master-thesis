```python
from gillespie.simulate import simulate
from gillespie.likelihood import log_p
import dask.array as da
from dask import delayed
import dask
from matplotlib import pyplot
```

```python
length = 10_000
```

```python
count = 100_000
batch = 100
```

```python
signals = []
for i in range(count//batch):
    sig = delayed(simulate)(batch, length, sig_network, initial_value=50, pure=False)
    signals.append(sig.persist())
```

```python
responses = []
for sig in signals:
    res = delayed(simulate)(batch, length, res_network, sig, initial_value=50, pure=False)
    responses.append(res.persist())
```

```python
likelihoods = []
traj_lengths = np.linspace(0, 7e3, 1000)
for sig, res in zip(signals, responses):
    lik = delayed(log_p)(traj_lengths, sig, res, res_network)
    likelihoods.append(lik)
```

```python
likelihood = da.concatenate([da.from_delayed(l, shape=(batch,1000), dtype=np.double) for l in likelihoods])
```

```python
pyplot.hist(likelihood[:,-1].compute(), bins='auto')
```

```python
conditional_entropy = da.mean(likelihood, axis=0).compute()
```

```python
other_res = []
for i in range(count//batch):
    sig = delayed(simulate)(1, length, sig_network, initial_value=50, pure=False)
    res = delayed(simulate)(1, length, res_network, sig, initial_value=50, pure=False)
    other_res.append(res.persist())
```

```python
len(other_res)
```

```python
import scipy
import scipy.special
```

```python
sig_t = da.concatenate([da.from_delayed(sig['timestamps'], shape=(batch, length), dtype=np.double) for sig in signals])
sig_c = da.concatenate([da.from_delayed(sig['components'], shape=(batch, 1, length), dtype=np.uint16) for sig in signals])
```

```python
sig_t = sig_t.rechunk((8000, -1)).persist()
sig_c = sig_c.rechunk((8000, -1, -1)).persist()
```

```python
traj_lengths = np.linspace(0, 7e3, 1000)
def parallel_log_p(st, sc, res):
    sig = {'timestamps': st, 'components': sc}
    return log_p(traj_lengths, sig, res, res_network)
```

```python
data = []
traj_lengths = np.linspace(0, 7e3, 1000)
for res in other_res:
    res_ll = da.blockwise(parallel_log_p, 'ij', sig_t, 'ik', sig_c, 'ilk', res, None, new_axes={'j': 1000}, dtype=np.double, concatenate=True)
#     res_ll = da.reduction(res_ll, scipy.special.logsumexp, scipy.special.logsumexp, axis=0, dtype=np.double).persist()
    data.append(res_ll)
data = da.concatenate(data)
```

```python
dist = data[:,-1].persist()
dist
```

```python
dist_range = [dist.min().compute(), dist.max().compute()]
```

```python
hist, bins = da.histogram(dist, bins=100, range=dist_range)
```

```python
pyplot.bar(x=(bins[:-1] + bins[1:])/2, width=np.diff(bins), height=hist.compute())
```

```python
marginal_entropy = da.reduction(data, scipy.special.logsumexp, scipy.special.logsumexp, axis=0, dtype=np.double).compute()
```

```python
me = -(marginal_entropy - np.log(data.shape[0]))
ce = -conditional_entropy
```
